```python
# Sample dataset: Concepts and corresponding target values
# concepts = [
#     ['sunny', 'warm', 'normal', 'strong', 'warm', 'same'],
#     ['sunny', 'warm', 'high', 'strong', 'warm', 'same'],
#     ['rainy', 'cold', 'high', 'strong', 'warm', 'change'],
#     ['sunny', 'warm', 'high', 'strong', 'cool', 'change']
# ]

concepts = [
    ['red', 'round', 'sweet'],
    ['green', 'round', 'sweet'],
    ['green', 'oval', 'lightly sweet'],
    ['red', 'round', 'sour'],
    ['green', 'round', 'sweet']
]

target = [True, True, False, True,True]  # True = Positive, False = Negative

# Function to print S and G at each step
def show(s, g, iteration):
    print(f"\nIteration {iteration}:")
    print(f"S: {s}")
    print(f"G: {g}")

# Candidate Elimination Algorithm (Simplified)
def candidate_elimination(concepts, target):
    # Step 1: Initialize S (most specific hypothesis)
    specific_h = concepts[0][:]  # Copy first positive example
    # Step 2: Initialize G (most general hypothesis)
    general_h = [['?' for _ in range(len(specific_h))]]

    show(specific_h, general_h, 0)

    # Step 3: Process each example one by one
    for i in range(len(concepts)):
        example = concepts[i]

        if target[i]:  # Positive example
            # Generalize specific_h to match this example
            for j in range(len(specific_h)):
                if example[j] != specific_h[j]:
                    specific_h[j] = '?'  # Generalize

            # Remove general hypotheses inconsistent with updated specific_h
            new_general_h = []
            for h in general_h:
                keep = False
                for k in range(len(specific_h)):
                    if h[k] == specific_h[k] or h[k] == '?':
                        keep = True
                    else:
                        keep = False
                        break
                if keep:
                    new_general_h.append(h)
            general_h = new_general_h

        else:  # Negative example
            # Specialize general_h to exclude this negative example
            new_general_h = []
            for h in general_h:
                for j in range(len(specific_h)):
                    if specific_h[j] != '?' and specific_h[j] != example[j]:
                        # Create a copy of h and specialize it
                        new_h = h[:]
                        new_h[j] = specific_h[j]
                        new_general_h.append(new_h)
            general_h = new_general_h

        # Print current S and G
        show(specific_h, general_h, i+1)

    return specific_h, general_h

# Run the algorithm
s_final, g_final = candidate_elimination(concepts, target)

# Print final hypotheses
print("\nFinal Specific Hypothesis (S):", s_final)
print("Final General Hypotheses (G):", g_final)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample data (you can replace this with your own)
X = np.array([
    [500, 20],
    [1500, 50],
    [700, 30],
    [2000, 70],
    [1000, 40],
    [250, 10],
    [1200, 60],
    [300, 15]
])

inertia_values = []

# Try k values from 1 to number of samples
for k in range(1, len(X)+1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)  # Sum of squared distances to closest centroid

# Plot inertia vs. k
plt.plot(range(1, len(inertia_values) + 1), inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.title('Elbow Method to Find Optimal k')
plt.show()
```
